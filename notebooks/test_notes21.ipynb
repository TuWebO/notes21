```python
from music.core import Note
from music.visualization import print_note_grid, plot_note_grid
```

```python
# Text based visualization
C = [Note("Cbb"), Note("Cb"), Note("C"), Note("C#"), Note("C##")]
D = [Note("Dbb"), Note("Db"), Note("D"), Note("D#"), Note("D##")]
E = [Note("Ebb"), Note("Eb"), Note("E"), Note("E#"), Note("E##")]
F = [Note("Fbb"), Note("Fb"), Note("F"), Note("F#"), Note("F##")]
G = [Note("Gbb"), Note("Gb"), Note("G"), Note("G#"), Note("G##")]
A = [Note("Abb"), Note("Ab"), Note("A"), Note("A#"), Note("A##")]
B = [Note("Bbb"), Note("Bb"), Note("B"), Note("B#"), Note("B##")]

notes = C + D + E + F + G + A + B
fig = print_note_grid(notes, "C")
fig = print_note_grid(notes, "G")
fig = print_note_grid(notes, "D")
fig = print_note_grid(notes, "A")
fig = print_note_grid(notes, "E")
fig = print_note_grid(notes, "B")
fig = print_note_grid(notes, "F#")
fig = print_note_grid(notes, "C#")
fig = print_note_grid(notes, "F")
fig = print_note_grid(notes, "Bb")
fig = print_note_grid(notes, "Eb")
fig = print_note_grid(notes, "Ab")
fig = print_note_grid(notes, "Gb")
fig = print_note_grid(notes, "Cb")
```
