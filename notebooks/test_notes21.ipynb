```python
from music.core import Note
from music.visualization import plot_note_grid_3d, animate_tonal_trajectory
from IPython.display import HTML
import matplotlib.pyplot as plt
```

```python
# 3D representation

notes = [
    # Low register
    Note("C", 3),
    Note("E", 3),
    Note("G", 3),
    
    # Mid register
    Note("C", 4),
    Note("E", 4),
    Note("G", 4),
    
    # Add chromatic tones
    Note("Bb", 4),
    Note("F#", 5),
    
    # High register melodic ascent
    Note("C", 5),
    Note("D", 5),
    Note("E", 5),
    Note("G", 6),
    ]

plot_note_grid_3d(notes, key="C", octave_range=(3, 6))
plt.show()
```

```python
# Example 1  The Golden Major Triad (C Major)
notes = [
    # C major triad across octaves
    Note("C", 3),
    Note("E", 3),
    Note("G", 3),

    Note("C", 4),
    Note("E", 4),
    Note("G", 4),

    Note("C", 5),
    Note("E", 5),
    Note("G", 5),
]

chart = plot_note_grid_3d(notes, key="C", octave_range=(3, 5))
```

```python
# Example 2  IVviIV Progression (Pop Golden Progression)
notes = [
    # C major (I)
    Note("C", 4), Note("E", 4), Note("G", 4),

    # G major (V)
    Note("G", 4), Note("B", 4), Note("D", 5),

    # A minor (vi)
    Note("A", 4), Note("C", 5), Note("E", 5),

    # F major (IV)
    Note("F", 4), Note("A", 4), Note("C", 5),
]

chart = plot_note_grid_3d(notes, key="C", octave_range=(4, 5))
```

```python
# Example 3  Add the Leading Tone Tension
notes = [
    Note("B", 4),
    Note("C", 5),
    Note("E", 5),
    Note("G", 5),
]

chart = plot_note_grid_3d(notes, key="C", octave_range=(4, 5))
```

```python
# Example 4  Add Chromatic Gold (Secondary Dominant)
notes = [
    # D7 chord
    Note("D", 4),
    Note("F#", 4),  # chromatic +1 column
    Note("A", 4),
    Note("C", 5),
]

chart = plot_note_grid_3d(notes, key="C", octave_range=(4, 5))
```

```python
notes = []

for octave in range(3, 7):
    notes += [
        Note("C", octave),
        Note("E", octave),
        Note("G", octave),
    ]

chart = plot_note_grid_3d(notes, key="C", octave_range=(3, 6))
```

```python
# Animations
#  Example 1  Golden Pop Progression Animation
notes = [
    # C
    Note("C", 4), Note("E", 4), Note("G", 4),
    # G
    Note("G", 4), Note("B", 4), Note("D", 5),
    # Am
    Note("A", 4), Note("C", 5), Note("E", 5),
    # F
    Note("F", 4), Note("A", 4), Note("C", 5),
]

ani = animate_tonal_trajectory(notes, key="C", octave_range=(4, 5))
play = HTML(ani.to_jshtml())
```

```python
# Example 2  Chromatic Tension Resolution
notes = [
    Note("C", 4),
    Note("E", 4),
    Note("F#", 4),  # tension
    Note("G", 4),
    Note("Bb", 4),  # tension
    Note("C", 5),   # resolution
]

ani = animate_tonal_trajectory(notes, key="C", octave_range=(4, 5))
play = HTML(ani.to_jshtml())
```
